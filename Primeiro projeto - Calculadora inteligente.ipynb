# **Calculadora**

```python
num1 = input('digite um número')
print(num1)
```

```python
num2 = input ('digite um número')
print(num2)
```

```python
convert_num1 = float(num1)
```

```python
convert_num2 = float(num2)
```

```python
# Solicitando a operação desejada
print("\nEscolha a operação:")
print("1 - Soma (+)")
print("2 - Subtração (-)")
print("3 - Multiplicação (*)")
print("4 - Divisão (/)")

opp = input("Digite o número da operação desejada: ")
```

```python
# Realizando a operação escolhida
if opp == "1":
    resultado = convert_num1 + convert_num2
    operacao = "Soma"
elif opp == "2":
    resultado = convert_num1 - convert_num2
    operacao = "Subtração"
elif opp == "3":
    resultado = convert_num1 * convert_num2
    operacao = "Multiplicação"
elif opp == "4":
    if convert_num2 == 0:
        resultado = "Erro! Divisão por zero."
        operacao = ""
    else:
        resultado = convert_num1 / convert_num2
        operacao = "Divisão"
else:
    resultado = "Opção inválida!"
    operacao = ""
```

```python
if operacao:
    print(f"\nResultado da {operacao}: {resultado}")
else:
    print(resultado)
```
